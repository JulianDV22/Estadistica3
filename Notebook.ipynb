```python
# Datos del ejercicio
data = [
    {"City": "Bogot", "GDP": 103.5, "Population": 7.18},
    {"City": "Medelln", "GDP": 44.1, "Population": 2.57},
    {"City": "Cali", "GDP": 22.4, "Population": 2.23},
    {"City": "Barranquilla", "GDP": 16.8, "Population": 1.23},
    {"City": "Cartagena", "GDP": 10.5, "Population": 1.03},
    {"City": "Bucaramanga", "GDP": 7.3, "Population": 0.58},
    {"City": "Pereira", "GDP": 6.2, "Population": 0.48},
    {"City": "Ccuta", "GDP": 5.1, "Population": 0.76},
    {"City": "Ibagu", "GDP": 4.8, "Population": 0.53},
    {"City": "Santa Marta", "GDP": 4.0, "Population": 0.52},
    {"City": "Manizales", "GDP": 3.8, "Population": 0.43},
    {"City": "Villavicencio", "GDP": 3.5, "Population": 0.5},
    {"City": "Pasto", "GDP": 3.2, "Population": 0.45},
    {"City": "Montera", "GDP": 3.0, "Population": 0.49},
    {"City": "Valledupar", "GDP": 2.8, "Population": 0.47},
    {"City": "Neiva", "GDP": 2.5, "Population": 0.35},
    {"City": "Popayn", "GDP": 2.3, "Population": 0.33},
    {"City": "Armenia", "GDP": 2.1, "Population": 0.3},
    {"City": "Sincelejo", "GDP": 2.0, "Population": 0.28},
    {"City": "Tunja", "GDP": 1.8, "Population": 0.25},
    {"City": "Florencia", "GDP": 1.7, "Population": 0.2},
    {"City": "Riohacha", "GDP": 1.5, "Population": 0.22},
    {"City": "Quibd", "GDP": 1.3, "Population": 0.13},
    {"City": "San Andrs", "GDP": 1.2, "Population": 0.08},
    {"City": "Yopal", "GDP": 1.1, "Population": 0.15},
    {"City": "Leticia", "GDP": 1.0, "Population": 0.05},
    {"City": "Arauca", "GDP": 0.9, "Population": 0.08},
    {"City": "Mocoa", "GDP": 0.8, "Population": 0.04},
    {"City": "Mit", "GDP": 0.7, "Population": 0.01},
    {"City": "Puerto Carreo", "GDP": 0.6, "Population": 0.01}
]

# Paso 1: Seleccionar 6 centroides iniciales aleatoriamente
import random
random.seed(42)  # Para que sea reproducible
centroids = random.sample(data, 6)

# Funcin para calcular la distancia euclidiana entre dos puntos
def euclidean_distance(point1, point2):
    return ((point1["GDP"] - point2["GDP"]) ** 2 + (point1["Population"] - point2["Population"]) ** 2) ** 0.5

# Asignar cada ciudad al cluster ms cercano
def assign_clusters(data, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for city in data:
        distances = [euclidean_distance(city, centroid) for centroid in centroids]
        closest_centroid = distances.index(min(distances))
        clusters[closest_centroid].append(city)
    return clusters

# Recalcular los centroides como el promedio de los puntos en cada cluster
def recalculate_centroids(clusters):
    new_centroids = []
    for cluster in clusters:
        if cluster:  # Evitar dividir por 0
            avg_gdp = sum(city["GDP"] for city in cluster) / len(cluster)
            avg_population = sum(city["Population"] for city in cluster) / len(cluster)
            new_centroids.append({"GDP": avg_gdp, "Population": avg_population})
    return new_centroids

# Ejecutar K-means manualmente (solo iteramos hasta que los centroides no cambien ms)
def kmeans(data, centroids, max_iterations=100):
    for _ in range(max_iterations):
        clusters = assign_clusters(data, centroids)
        new_centroids = recalculate_centroids(clusters)
        
        # Si los centroides no cambian, detener el algoritmo
        if new_centroids == centroids:
            break
        centroids = new_centroids
    
    return clusters, centroids

# Ejecutar el algoritmo
clusters, final_centroids = kmeans(data, centroids)

# Imprimir resultados
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1}:")
    for city in cluster:
        print(f"  {city['City']} - GDP: {city['GDP']}, Population: {city['Population']}")
    print()

print("Centroides finales:")
for i, centroid in enumerate(final_centroids):
    print(f"  Centroid {i+1}: GDP: {centroid['GDP']}, Population: {centroid['Population']}")
```

```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Datos de ejemplo
data = np.array([
    [1, 2],
    [2, 3],
    [3, 4],
    [5, 6],
    [8, 9],
    [10, 11]
])

# Calcular la matriz de distancias
distance_matrix = sch.distance.pdist(data, metric='euclidean')

# Crear el linkage matrix usando el mtodo de enlace completo
linkage_matrix = sch.linkage(distance_matrix, method='complete')

# Crear y visualizar el dendograma
plt.figure(figsize=(10, 7))
sch.dendrogram(linkage_matrix)
plt.title('Dendrograma usando el mtodo de enlace completo')
plt.xlabel('ndice de muestra')
plt.ylabel('Distancia')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Datos
data = {
    'City': ['Bogot', 'Medelln', 'Cali', 'Barranquilla', 'Cartagena', 'Bucaramanga (test)', 'Pereira', 'Ccuta (test)', 
             'Ibagu (test)', 'Santa Marta', 'Manizales', 'Villavicencio', 'Pasto', 'Montera', 'Valledupar', 'Neiva', 
             'Popayn', 'Armenia', 'Sincelejo', 'Tunja', 'Florencia', 'Riohacha', 'Quibd', 'San Andrs', 'Yopal', 
             'Leticia', 'Arauca (test)', 'Mocoa', 'Mit', 'Puerto Carreo (test)'],
    'GDP': [103.5, 44.1, 22.4, 16.8, 10.5, 7.3, 6.2, 5.1, 4.8, 4, 3.8, 3.5, 3.2, 3, 2.8, 2.5, 2.3, 2.1, 2, 1.8, 
            1.7, 1.5, 1.3, 1.2, 1.1, 1, 0.9, 0.8, 0.7, 0.6],
    'Population': [7.18, 2.57, 2.23, 1.23, 1.03, 0.58, 0.48, 0.76, 0.53, 0.52, 0.43, 0.5, 0.45, 0.49, 0.47, 0.35, 
                   0.33, 0.3, 0.28, 0.25, 0.2, 0.22, 0.13, 0.08, 0.15, 0.05, 0.08, 0.04, 0.01, 0.01]
}

df = pd.DataFrame(data)

# Extraer las variables de inters
X = df[['GDP', 'Population']].values

# 1. Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Calcular PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

# Resultados de PCA
explained_variance = pca.explained_variance_ratio_
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_

# 3. Matriz de covarianza
cov_matrix = np.cov(X_scaled.T)

# Resultados
print("1. Matriz de Covarianza:\n", cov_matrix)
print("\n2. Eigenvalues:\n", eigenvalues)
print("\n3. Varianza explicada por el eigenvalue:\n", explained_variance)
print("\n4. Eigenvector:\n", eigenvectors)
print("\n5. Matriz Proyectada:\n", X_pca)

# 6. Calcular el error de reconstruccin
X_reconstructed = pca.inverse_transform(X_pca)
reconstruction_error = np.mean((X_scaled - X_reconstructed) ** 2)
print("\n6. Error o diferencia entre la matriz proyectada:\n", reconstruction_error)

# 7. Pintar todas las ciudades en 1 dimensin
plt.figure(figsize=(12, 6))
plt.scatter(X_pca, np.zeros_like(X_pca), c='blue', label='Datos proyectados en 1D')
plt.title('Datos en 1 Dimensin (Proyeccin PCA)')
plt.xlabel('Componente Principal 1')
plt.yticks([])
plt.legend()
plt.show()

# 8. Pintar todas las ciudades en 2 dimensiones
plt.figure(figsize=(12, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c='blue', label='Datos originales en 2D')
plt.title('Datos Originales en 2 Dimensiones')
plt.xlabel('GDP (Estandarizado)')
plt.ylabel('Population (Estandarizado)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('amazon_products.csv')

# Mostrar las primeras filas del dataframe para entender su estructura
print(df.head())

# Crear variables dummy para las variables categricas
df_dummies = pd.get_dummies(df, drop_first=True)

# Mostrar las primeras filas del dataframe transformado
print(df_dummies.head())

# Calcular la nueva matriz de correlacin
correlation_matrix_dummies = df_dummies.corr()

# Visualizar la matriz de correlacin con las variables dummy
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_dummies, annot=False, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlacin con Variables Dummy')
plt.show()

# Encontrar la variable categrica con mayor correlacin con 'sales_volume'
correlations_dummies = correlation_matrix_dummies['sales_volume'].sort_values(ascending=False)
print('Variables dummy con mayor correlacin con sales_volume:')
print(correlations_dummies)

```
